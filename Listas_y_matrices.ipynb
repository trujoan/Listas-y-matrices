```python
#Programa para una lista
#Def nos sirve para definir una función o un bloque que de código que después se puede reutilizar, simplemente haciéndole un llamado.

def crear_lista():
    lista = []
    pedir_datos=True
    while pedir_datos==True:
      elemento=input("Ingrese el elemento que desea agregar a la lista o presione x para terminar: ")
      #El método append, permite agregar elementos a una lista.
      if elemento!="x":
        lista.append(elemento)
      else:
        pedir_datos=False
        return lista

frutas=crear_lista()

print(frutas)

#La función crear lista pide elementos al usuario que se agregarán a la lista, una vez que el usuario digite x, el programa imprimirá la lista completa.
```

```python
#Imprime uno a uno los elementos de una lista, la función debe recibir la lista o el nombre de una lista que se creó previamente.
def recorrer_lista(lista):
  for elemento in lista:
    print(elemento)

recorrer_lista(frutas)
#recorrer_lista(nombres)
```

```python
def agregar_elemento(lista, elemento):
  lista.append(elemento)
  return lista

agregar_elemento(nombres, "José")
```

```python
def modificar_elemento(lista, indice, nuevo_elemento):
  lista[indice]=nuevo_elemento
  return lista

modificar_elemento(nombres, 0, "Camilo")
modificar_elemento(frutas, 2, "Limón")

print(nombres)
print(frutas)
```

```python
#Esta función elimina el elemento que se indique de la lista, se debe ingresar la lista o el nombre de la lista si ya ha sido creada y el indice de ubicación del elemento.
def extraer_elemento(lista, indice):
  elemento=lista[indice]
  print (f"Se eliminará de la lista el elemento en la posicion {indice} que corresponde a {lista[indice]}")
  lista.remove(elemento)
  return lista

print(nombres)
print(extraer_elemento(nombres, -1))
```

```python
def crear_matriz():
  filas=int(input("Ingrese el número de filas de la matriz: "))
  columnas=int(input("Ingrese el número de columnas de la matriz: "))
  matriz=[]
  for i in range(filas):
    matriz.append([])
    for j in range(columnas):
      elemento=input(f"Ingrese el elemento en la posición [{i}][{j}]: ")
      matriz[i].append(elemento)
  return matriz

prueba=crear_matriz()

print(prueba)
```

```python
def recorrer_matriz(matriz):
  filas=len(matriz)
  columnas=len(matriz[0])
  for i in range(filas):
    for j in range(columnas):
      print(f"Elemento en la posición [{i}][{j}]: {matriz[i][j]}")

recorrer_matriz(prueba)
```

```python
def modificar_elemento_matriz(matriz, fila, columna, elemento_nuevo):
  matriz[fila][columna]=elemento_nuevo
  return matriz

modificar_elemento_matriz(prueba, 0, 1, "Hola")
```

```python
def eliminar_elemento_matriz(matriz, fila, columna):
  elemento=matriz[fila][columna]
  print(f"Se eliminará el elemento en la posición [{fila}][{columna}]: {elemento}")
  matriz[fila][columna]=None
  return matriz

matriz_gaby=eliminar_elemento_matriz(prueba, 0, 1)
print(matriz_gaby)
```
